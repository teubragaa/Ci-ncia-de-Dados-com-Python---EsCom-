# A Plant-Based Coffee Shop

### Loading the Data

Activity Goals:
- Import pandas.
- Load the coffee survey data.
- Display the dataframe.

```python
import pandas as pd

survey = pd.read_csv('data/coffee-survey-results.csv')
survey
```

### Narrowing down the columns

Activity Goal:
- Display all the column names.

```python
survey.columns
```

Activity Goals:
- Select the needed columns from the survey.
- Assign these columns to a new dataframe named dairy
- Display new dataframe.

```python
needed_columns = [
    "What kind of dairy? (Whole milk)",
    "What kind of dairy? (Skim milk)",
    "What kind of dairy? (Half and half)",
    "What kind of dairy? (Coffee creamer)",
    "What kind of dairy? (Flavored creamer)",
    "What kind of dairy? (Oat milk)",
    "What kind of dairy? (Almond milk)",
    "What kind of dairy? (Soy milk)",
]
dairy = survey[needed_columns]
dairy 
```

### Renaming the columns



Activity Goals:
- Rename the columns using the name map.
- Display the result.

```python
name_map = {
    'What kind of dairy? (Whole milk)': 'Whole milk',
    'What kind of dairy? (Skim milk)': 'Skim milk',
    'What kind of dairy? (Half and half)': 'Half and half',
    'What kind of dairy? (Coffee creamer)': 'Coffee creamer',
    'What kind of dairy? (Flavored creamer)': 'Flavored creamer',
    'What kind of dairy? (Oat milk)': 'Oat milk',
    'What kind of dairy? (Almond milk)': 'Almond milk',
    'What kind of dairy? (Soy milk)': 'Soy milk',
}
```

```python
dairy = dairy.rename(columns=name_map)
dairy #display the dataframe
```

### Making sense of the data

Activity Goals:
- Create a boolean dataframe with True for NaNs and False otherwise.
- Add up the True values per column to get the number of NaNs.

```python
dairy.isna().sum()
```

Activity Goals:
- Remove rows with NaN values.
- Display the remaining rows.

```python
dairy = dairy.dropna()
dairy 
```

### What dairy options are most preferred?


Activity Goals:
- Calculate the percentage for each dairy.
- Display the result.

```python
dairy_preferences = dairy.mean() * 100
dairy_preferences #display results
```

### Visualizing dairy preferences

Activity Goals:
- Sort diary_preferences.
- Display the series.

```python
dairy_preferences = dairy_preferences.sort_values()
dairy_preferences 
```


Activity Goals:
- Import matplotlib.
- Make a horizontal bar chart of dairy preferences.
- Label the x-axis.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 3))
plt.barh(dairy_preferences.index, dairy_preferences)
plt.xlabel("Percent")
```

1. Which sweeteners should your coffee shop stockpile to meet customer preferences?

Primeiro preciso conferir as colunas

```python
survey.columns
```

Apos saber quais colunas tenho, seleciono-as de forma precisa

```python
sweetener_columns = [
    'What kind of sugar or sweetener? (Granulated Sugar)',
    'What kind of sugar or sweetener? (Artificial Sweetener)',
    'What kind of sugar or sweetener? (Honey)',
    'What kind of sugar or sweetener? (Maple Syrup)',
    'What kind of sugar or sweetener? (Stevia)',
    'What kind of sugar or sweetener? (Agave Nectar)',
    'What kind of sugar or sweetener? (Brown Sugar)',
    'What kind of sugar or sweetener? (Raw Sugar)',
]

sweeteners = survey[sweetener_columns]
sweeteners
```

Agora eu renomeio para facilitar o entendimento

```python
sweetener_columns_map = {
    'What kind of sugar or sweetener? (Granulated Sugar)' : 'Granulated Sugar' ,
    'What kind of sugar or sweetener? (Artificial Sweetener)' : 'Artificial Sweetener',
    'What kind of sugar or sweetener? (Honey)' : 'Honey',
    'What kind of sugar or sweetener? (Maple Syrup)' : 'Maple Syrup',
    'What kind of sugar or sweetener? (Stevia)' : 'Stevia',
    'What kind of sugar or sweetener? (Agave Nectar)' : 'Agave Nectar',
    'What kind of sugar or sweetener? (Brown Sugar)' : 'Brown Sugar',
    'What kind of sugar or sweetener? (Raw Sugar)' : 'Raw Sugar',
}

sweeteners = sweeteners.rename(columns=sweetener_columns_map)
sweeteners
```

Retiro as linhas com NaN

```python
sweeteners.dropna()
```

Vou somar para ver quais sao as mais utilizadas para que consiga organizar por nivel de prioridade o estoque.

```python
sweeteners.sum().sort_values(ascending=False)
```

```python
sweeteners_totals = sweeteners.sum().sort_values(ascending=True)

plt.figure(figsize=(6,2))
plt.barh(sweeteners_totals.index, sweeteners_totals)
plt.xlabel("Número de Votos")
plt.title("Preferência de Adoçantes")
```

Tendo em vista o grafico acima, é uma boa pratica colocar estoque nos 3 adocantes acima.

2. Which brewing methods should your coffee shop offer to optimize customer satisfaction?

```python

brew_columns = [
    'How do you brew your coffee? (French press)',
    'How do you brew your coffee? (Espresso)',
    'How do you brew your coffee? (Coffee brewing machine)',
    'How do you brew your coffee? (Pod/capsule machine)',
    'How do you brew your coffee? (Instant coffee)',
    'How do you brew your coffee? (Bean-to-cup machine)',
    'How do you brew your coffee? (Cold brew)',
    'How do you brew your coffee? (Coffee extract)'
]

brewing = survey[brew_columns]
brewing
```

```python
brew_columns_map = {
    'How do you brew your coffee? (French press)' : '(French press)',
    'How do you brew your coffee? (Espresso)': 'Espresso',
    'How do you brew your coffee? (Coffee brewing machine)' : 'Coffe brewing machine',
    'How do you brew your coffee? (Pod/capsule machine)' : 'Pod/capsule machine',
    'How do you brew your coffee? (Instant coffee)' : 'Instant coffee',
    'How do you brew your coffee? (Bean-to-cup machine)': 'Bean-to-cup machine',
    'How do you brew your coffee? (Cold brew)' : 'Cold Brew',
    'How do you brew your coffee? (Coffee extract)' : 'Coffe Extract',
}

brewing = brewing.rename(columns=brew_columns_map)
```

```python
brewing.dropna()
brewing.sum().sort_values(ascending=False)
```

```python
soma_brewing = brewing.sum().sort_values(ascending=True)

plt.figure(figsize=(6,2))
plt.barh(soma_brewing.index, soma_brewing)
```
